# Questão 20

Crie um código para preencher os valores faltantes de glucose utilizando como valor, a media dos valores de 'Glucose' separada por grupo de idade, opu seja, se a pessoa sem dado de "Glucose" tiver 21 anos utilize media de Glucose de 21 anos para preencher o valor faltante.

Apos o preenchimento dos dados faltantes, selecione todas as alternativas abaixo onde temos um par de "Idade"(Age)
e "Glucose Media" utilizados para o preenchimento de ao menos um valor faltante.

Por exemplo, se no conjunto de daos tivermos um paciente com 33 anos sem dados de Glucose, selecione a opcao onde temos "idade: 33" e onde a "Glucose media" é igual a "Glucose média"para as pessoas de 33 anos.

Obs: Valores foram truncados, desconsiderando as casas decimais.

Resposta:
Idade: 21 Glucose média: 110
Idade: 22 Glucose média: 111
Idade: 37 Glucose média: 137
Idade: 41 Glucose média: 135

```python
#20
#Printar a idade media
dados_1 = df[df['Glucose'].isnull()]
idade_2 = dados_1['Age']
for x in idade_2:
    dados_aux = df[df['Age'] == x]
    z = dados_aux['Glucose'].mean()
    print(x,' ',z)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
df= pd.read_csv('diabetes.csv')
```

```python
df['Glucose'] = df['Glucose'].replace(0, np.NaN)
```

```python
df.head()
```

```python
df.Outcome.value_counts(normalize = True)*100
```

```python
#y é a variável target, e X é a variavel teste
y = df['Outcome']
X = df.drop('Outcome', axis = 1)
```

```python
X
```

```python
#stratify indica que quero a mesma proporçao de y da linha 148
X_train, X_test,y_train,y_test = train_test_split(X,y,
                                                  test_size=0.4,
                                                  random_state=0,
                                                  stratify=y)
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
#escalonando todos os dados de treino e teste em 0 e 1
X_train[X_train.columns] = scaler.transform(X_train)
X_test[X_test.columns] = scaler.transform(X_test)
```
